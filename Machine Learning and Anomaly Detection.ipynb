```python
# Import the required function from sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Extract Response and Predictors
y = pd.DataFrame(reg_data['Total'])
X = pd.DataFrame(reg_data['Year'])

linreg = LinearRegression()

# Split the Dataset into random Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

# Check the sample sizes
print("Train Set :", X_train.shape, y_train.shape)
print("Test Set  :", X_test.shape, y_test.shape)
```

```python
linreg.fit(X_train, y_train)
```

```python
# Formula for the Regression line
regline_x = X_train
regline_y = linreg.intercept_ + linreg.coef_ * X_train

# Plot the Linear Regression line
f, axes = plt.subplots(1, 1, figsize=(16, 8))
plt.scatter(X_train, y_train)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()
```

```python
print("Explained Variance (R^2) \t:", linreg.score(X_train, y_train))
# Import the required metric from sklearn
from sklearn.metrics import mean_squared_error

# Predict the response on the train set
y_train_pred = linreg.predict(X_train)

# Compute MSE on the train set
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_pred))
```

```python
# Compute MSE on the test set
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_pred))
```

However, we see that the correlation jointplot of the predictor against the response seems to hint at a non-linear relationship between the two. We now look at non-linear (polynomial) regression to see if it leads to a better model.

```python
# Polynomial Regression

from sklearn.preprocessing import PolynomialFeatures

# Extract Response and Predictors
y = pd.DataFrame(reg_data['Total'])
X = pd.DataFrame(reg_data['Year'])

linreg = LinearRegression()

# Split the Dataset into random Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)


# Function to perform polynomial regression.
def poly_reg(d):
    poly_features = PolynomialFeatures(degree = d)
    poly_X_train = poly_features.fit_transform(X_train)
    linreg.fit(poly_X_train, y_train)
    y_train_pred = linreg.predict(poly_X_train)
    y_test_pred = linreg.predict(poly_features.fit_transform(X_test))
    print("Mean Squared Error (MSE) on the train set\t:", mean_squared_error(y_train, y_train_pred))
    print("Mean Squared Error (MSE) on the test set\t:", mean_squared_error(y_test, y_test_pred))
    f = plt.figure(figsize=(16, 8))
    plt.title("Train Set")
    plt.scatter(X_train, y_train, color = "green")
    plt.scatter(X_train, y_train_pred, color = "red")
    plt.show()
    f = plt.figure(figsize=(16, 8))
    plt.title("Test Set")
    plt.scatter(X_test, y_test, color = "green")
    plt.scatter(X_test, y_test_pred, color = "red")
    plt.show()
```

The optimal model is one with low bias and low variance. The following code shows how data is fit on polynomials of different degrees (ranging from 1 to the total number of points in train set). This helps visualize the bias-variance trade off, i.e; a low variance can be achieved by increasing the bias and vice versa.

```python
count=22

colors = iter(plt.cm.rainbow(np.linspace(0, 1, count)))

fig, axs = plt.subplots(2, 1, figsize=(20, 20))
'''axs[0, 0].set_title('Train Set Predictions')
axs[1, 0].set_title('Test Set Predictions')'''
axs[0].set_title('Train Set MSE')
axs[1].set_title('Test Set MSE')

'''axs[0, 0].set_xlabel("X train")
axs[0, 0].set_ylabel("y")
axs[0, 0].set_ylim([-3, 3])'''

'''axs[1, 0].set_xlabel("X test")
axs[1, 0].set_ylabel("y")
axs[1, 0].set_ylim([-3, 3])'''

axs[0].set_xlabel("polynomial degree")
axs[0].set_ylabel("train MSE")

axs[1].set_xlabel("polynomial degree")
axs[1].set_ylabel("test MSE")

#axs[0,0].scatter(X_train, y_train, c = 'k', label = "y train")
#axs[1,0].scatter(X_test, y_test, c = 'k', label = "y test")

train_mse = [0 for i in range(22)]
test_mse = [0 for i in range(22)]

for i in range(count):
    c = next(colors)
    
    poly_features = PolynomialFeatures(degree = i+1)
    poly_X_train = poly_features.fit_transform(X_train)
    linreg.fit(poly_X_train, y_train)
    y_train_pred = linreg.predict(poly_X_train)
    y_test_pred = linreg.predict(poly_features.fit_transform(X_test))

    
    '''axs[0,0].scatter(X_train,
                     y_train_pred,
                     color = c,
                     label = "deg: {}".format(i+1))
  
    axs[1,0].scatter(X_test,
                    y_test_pred,
                    color = c,
                    label = "deg: {}".format(i+1))'''
    
    
    iter_train_mse = mean_squared_error(y_train_pred, y_train)
    iter_test_mse = mean_squared_error(y_test_pred, y_test)

    train_mse[i]=iter_train_mse
    test_mse[i] = iter_test_mse
    
    # plot train and test MSE of i-th degree polynomial fit
    axs[0].plot(i+1,
                iter_train_mse,
                color = c,
                label = "deg: {}".format(i+1),
                marker = 'D',
                markersize = 12,
                markeredgecolor = 'black',
                markeredgewidth = 3)
  
    axs[1].plot(i+1,
                iter_test_mse,
                color = c,
                label = "deg: {}".format(i+1),
                marker = 'D',
                markersize = 12,
                markeredgecolor = 'black',
                markeredgewidth = 3)

    #plot dashed line to interpolate MSE measures
    axs[0].plot(range(1,count+1), train_mse, 'k--')
    axs[1].plot(range(1,count+1), test_mse, 'k--')

    # draw legends
    axs[0].legend(loc = "upper right",
                bbox_to_anchor = (1.2, 1.01),
                prop = {'size': 12})
    axs[1].legend(loc = "upper right",
                bbox_to_anchor = (1.2, 1.01),
                prop = {'size': 12})
```
