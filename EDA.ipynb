#### Which country consumes how much of which renewable energy?

```python
def sort_display(datasetName, colName):
    """
    Function to sort and display a given dataset according to a specific variable. 
    Additionally, it displays the barplot of the variable against 'Country'
    
    """
    datasetName = datasetName.sort_values(by=[colName], ascending=False)
    datasetName.head()
    
    plt.xticks(rotation=90)
    sb.barplot(data=datasetName, x=datasetName["Country"], y=datasetName[colName])
```

```python
sort_display(re_eng_countries, 'Hydro(TWh)')
```

Here we see that China is the lead consumer of Hydro energy, followed by Canada and Brazil. 

```python
sort_display(re_eng_countries, 'Biofuel(TWh)')
```

Here we see that China is the lead consumer of BioFuel energy, followed by USA and Germany. 

```python
sort_display(re_eng_countries, 'Solar PV (TWh)')
```

Here we see that China is the lead consumer of Solar PV energy, followed by USA and Brazil. 

```python
sort_display(re_eng_countries, 'Geothermal (TWh)')
```

Here we see that USA is the lead consumer of Geothermal energy, followed by Indonesia and Turkey. 

```python
sort_display(re_eng_countries, 'Total (TWh)')
```

Here we see that China is the lead consumer of energy, followed by USA and Brazil. 

### Renewable sources vs Non-Renewable sources

#### Which energy source comsumption is more energy?

```python
re_vs_nonre = {}
re_vs_nonre['Non-Renewable'] = nonre_eng['Contribution (TWh)'].sum()
re_vs_nonre['Renewable'] = re_eng_tot_gen['Contribution (TWh)'].sum()
sources = list(re_vs_nonre.keys())
energy_consump = list(re_vs_nonre.values())

fig = plt.figure(figsize = (5, 5))
 
# creating the bar plot
plt.bar(sources, energy_consump)
 
plt.xlabel("Source")
plt.ylabel("Consumption (TWh)")
plt.title("Renewable vs Non-Renewable Energy Total Consumption")
plt.show()
```

#### Global Total Energy Consumption vs Renewable Energy Generation Vs Non Renewable Energy Generation

```python
#First we create a DataFrame with Total, Renewable, and Non-Renewable energy Generation
global_energy_use = {}
years = []

for i in range(1990, 2005):
    years.append(i)
        
global_energy_use['Year'] = years
global_energy_use_df= pd.DataFrame(global_energy_use)

global_energy_use_df['World'] = cont_consump['World']


data_sum = pd.DataFrame.sum(re_eng_gen_90_17[['Hydro(TWh)', 'Biofuel(TWh)', 'Solar PV (TWh)', 'Geothermal (TWh)']],axis=1)

global_energy_use_df['total_renew_energy.Twh'] = data_sum

global_energy_use_df['total_non_renew_energy.Twh'] = global_energy_use_df['World'] - global_energy_use_df['total_renew_energy.Twh']

global_energy_use_df
```

```python
#plot both time series
plt.plot(global_energy_use_df['Year'], global_energy_use_df['total_renew_energy.Twh'], label='total_renew_energy', linewidth=3)
plt.plot(global_energy_use_df['Year'], global_energy_use_df['total_non_renew_energy.Twh'], color='red', label='total_non_renew_energy', linewidth=1.5)
plt.plot(global_energy_use_df['Year'], global_energy_use_df['World'], color='green', label='world', linewidth=1.5)


#add title and axis labels
plt.title('Global Total vs Renewable Vs Non Renewable Energy Generation')
plt.xlabel('Year')
plt.ylabel('Consumption (TWh)')

#add legend
plt.legend()

plt.legend(bbox_to_anchor=(1.05, 1))

#display plot
plt.show()
```
